```python
from sklearn.preprocessing import StandardScaler

X = df[features].values
target_column = 'Botnet'

y = df[target_column].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.feature_selection import RFECV
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

estimator = DecisionTreeClassifier(random_state=41, max_depth=20)

cv = StratifiedKFold(n_splits=3)

rfecv = RFECV(estimator=estimator, step=1, cv=cv, scoring='accuracy', n_jobs=-1)

rfecv.fit(X_scaled, y)


print("Optimal number of features:", rfecv.n_features_)
print("Selected features:", [index for index, selected in enumerate(rfecv.support_) if selected])
```

Firstly , we'll do Train test split to split our data into Training and testing data

```python
import numpy
from sklearn.model_selection import train_test_split

numpy.random.seed(1111)

(training_inputs,
testing_inputs,
training_classes,
testing_classes)=train_test_split(features_values, classes, train_size=0.75, random_state=1)
```

DECISION TREES

```python
from sklearn.tree import DecisionTreeClassifier

clf= DecisionTreeClassifier(random_state=1)

clf.fit(training_inputs, training_classes)

from sklearn.metrics import precision_score, f1_score, accuracy_score, recall_score

# Make predictions on the testing data
predictions = clf.predict(testing_inputs)

# Calculate precision, F1 score, accuracy, and recall
precision = precision_score(testing_classes, predictions, average='weighted')
f1 = f1_score(testing_classes, predictions, average='weighted')
accuracy = accuracy_score(testing_classes, predictions)
recall = recall_score(testing_classes, predictions, average='weighted')

print("Precision:", precision)
print("F1 Score:", f1)
print("Accuracy Score:", accuracy)
print("Recall:", recall)
```

```python
from sklearn.metrics import confusion_matrix

predicted_classes = clf.predict(testing_inputs)

confusion = confusion_matrix(testing_classes, predicted_classes)

print("Confusion Matrix:")
print(confusion)
```

An approximate of 80% , Notbad... NOW WE CAN TRY WITH RANDOM FORESTS

IMPLEMENTING XGBOOST..


```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_values, classes, test_size=0.2, random_state=42)
```

```python
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)
```

```python
params = {
    'objective': 'multi:softmax',  # For multi-class classification
    'num_class': 28,      # Specify the number of classes
    'max_depth': 3,
    'learning_rate': 0.1,
    'subsample': 0.8,
    'colsample_bytree': 0.8,
    'seed': 42
}
```

Train the xgboost model

```python
num_rounds = 100  # Number of boosting rounds
xgbmodel = xgb.train(params, dtrain, num_rounds)
```

Results

```python
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score

# Make predictions on the test data using the XGBoost model
predictions = xgbmodel.predict(dtest)

# Calculate classification metrics
accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions, average='weighted')
f1 = f1_score(y_test, predictions, average='weighted')
recall = recall_score(y_test, predictions, average='weighted')

print("Accuracy:", accuracy)
print("Precision (Weighted):", precision)
print("F1 Score (Weighted):", f1)
print("Recall (Weighted):", recall)
```

IMPLEMENTING LGB

```python
import lightgbm as lgb
```

Set the hyperparameters for the LightGBM model

```python
train_data = lgb.Dataset(X_train, label=y_train, free_raw_data=False)
test_data = lgb.Dataset(X_test, label=y_test, free_raw_data=False)
```

```python
params = {
    'objective': 'multiclass',        # Multi-class classification task
    'num_class': 28,                 # Specify the number of classes (28 in this case)
    'metric': 'multi_logloss',       # Multi-class log loss metric
    'num_leaves': 31,                # Maximum number of leaves in each tree
    'learning_rate': 0.1,           # Learning rate
    'feature_fraction': 0.9,        # Subsample ratio of columns when constructing each tree
    'bagging_fraction': 0.8,        # Subsample ratio of the training instances
    'bagging_freq': 5,              # Frequency of bagging
    'seed': 42                      # Random seed for reproducibility
}
```

Train The Model

```python
num_rounds = 100

# Train the LightGBM model
model = lgb.train(params, train_data, num_rounds, valid_sets=[train_data, test_data])
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score

# Make predictions on the test data
y_pred = model.predict(X_test, num_iteration=model.best_iteration)

# Convert the predicted probabilities to class labels
y_pred_class = np.argmax(y_pred, axis=1)

# Calculate accuracy, precision, F1-score, and recall
accuracy = accuracy_score(y_test, y_pred_class)
precision = precision_score(y_test, y_pred_class, average='weighted')
f1 = f1_score(y_test, y_pred_class, average='weighted')
recall = recall_score(y_test, y_pred_class, average='weighted')

# Print the metrics
print("Accuracy:", accuracy)
print("Precision (weighted):", precision)
print("F1 Score (weighted):", f1)
print("Recall (weighted):", recall)
```
